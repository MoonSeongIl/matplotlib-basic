## matplotlib library
    
   - plt.subplot2grid 사용

## practice (구현하기)

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(10,10))
ax1 = plt.subplot2grid((3,3),(0,0),fig=fig,colspan=3)
ax2 = plt.subplot2grid((3,3),(1,0),fig=fig,colspan=2)
ax3 = plt.subplot2grid((3,3),(1,2),fig=fig,rowspan=2)
ax4 = plt.subplot2grid((3,3),(2,0),fig=fig)
ax5 = plt.subplot2grid((3,3),(2,1),fig=fig)
```
